## Part 1.5 a - Experiment with various parameters to improve performance of the base model

#### The n_neighbour parameter

```python
rang=np.arange(1,30)

ac_scores = []

# perform accuracy metrics for values from 1,2,3,....30
for k in rang:
    knn_a = KNeighborsClassifier(n_neighbors=k,)
    knn_a.fit(x_train, y_train)
    # predict the response
    y_pred = knn_a.predict(x_test)
    # evaluate accuracy
    scores = accuracy_score(y_test, y_pred)
    ac_scores.append(scores)
    print('for the k = {} value the testing  {}'.format(k,round(ac_scores[-1],2)))

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = rang[MSE.index(min(MSE))]

print("The optimal number of neighbors is %d" % optimal_k)
```

```python
knn_n=KNeighborsClassifier(n_neighbors=optimal_k)
knn_n.fit(x_train,y_train)
print("The accuracy Score when n_neighbour is 1 : %1.2f " %knn_n.score(x_test,y_test))
```

```python
print(classification_report(y_test,knn_n.predict(x_test)))
```

```python
plt.figure(figsize=(10,7))
plt.plot(rang,ac_scores,"*-")
plt.title("Testing Accuracy Scores");
```

```python
plt.plot(rang, MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```

### Weights parameter

```python
ac_scores_w = []
weights= ['uniform','distance']
for w in weights:
    knn_w = KNeighborsClassifier(weights=w)
    knn_w.fit(x_train, y_train)
    # predict the response
    y_pred_w = knn_w.predict(x_test)
    scores = accuracy_score(y_test, y_pred_w)
    ac_scores_w.append(scores)
    print('for the weights = {}  and testing  {}'.format(w,round(ac_scores[-1],2)))
```

### When n_neighbor = 1 and weight = distance

```python
knn_w=KNeighborsClassifier(n_neighbors=1,weights='distance')
knn_w.fit(x_train,y_train)
print("The accuracy Score when n_neighbour is 1 and weights= distance is  : %1.2f " %knn_w.score(x_test,y_test))
```

When neighbors is one and weights is distance there is no change in the accuracy

### Algoritham parameter

```python
ac_scores = []
tr_scores = []
algor= ['auto','ball_tree','kd_tree','brute']
for a in algor:
    knn = KNeighborsClassifier(n_neighbors=1,algorithm=a)
    knn.fit(x_train, y_train)
    # predict the response
    y_pred = knn.predict(x_test)
    scores = accuracy_score(y_test, y_pred)
    ac_scores.append(scores)
    s = knn.score(x_train,y_train)
    tr_scores.append(s) 
    print('for the algoritham =  {}  then the trainig  {}  and testing  {}'.format(a,round(tr_scores[-1],2),round(ac_scores[-1],2)))
```

- There is no change is observed in accuracy when algoritham is changes

### Metric parameter

```python
ac_scores = []
tr_scores = []
metric = ['minkowski','euclidean','manhattan']
for a in metric:
    knn = KNeighborsClassifier(n_neighbors=1,metric=a)
    knn.fit(x_train, y_train)
    # predict the response
    y_pred = knn.predict(x_test)
    scores = accuracy_score(y_test, y_pred)
    ac_scores.append(scores)
    s = knn.score(x_train,y_train)
    tr_scores.append(s) 
    print('for the metric = {} value the trainig  {}  and testing  {}'.format(a,round(tr_scores[-1],2),round(ac_scores[-1],2)))
```

```python
ac_scores = []
tr_scores = []
p = [1,2,3,4,5]
for a in p:
    knn = KNeighborsClassifier(n_neighbors=1,p=a)
    knn.fit(x_train, y_train)
    # predict the response
    y_pred = knn.predict(x_test)
    scores = accuracy_score(y_test, y_pred)
    ac_scores.append(scores)
    s = knn.score(x_train,y_train)
    tr_scores.append(s) 
    print('for the p = {} value the trainig  {}  and testing  {}'.format(a,round(tr_scores[-1],2),round(ac_scores[-1],2)))
MSE = [1 - x for x in ac_scores]
M = [1 - x for x in tr_scores]

# determining best k
optimal_k = rang[MSE.index(min(MSE))]
opt_k = rang[M.index(min(M))]
print("The optimal number of p is %d" % optimal_k)
print("The optimal number of p is %d" % opt_k)
```

- There is a change is observed in accuracy when p=3

```python
plt.plot(p,MSE)
plt.xlabel('p value')
plt.ylabel('Misclassification Error')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(p,ac_scores,"*-")
plt.title("Testing Accuracy Scores")
plt.xlabel("p value")
plt.ylabel("Accuracy");
```

```python
knn_p = KNeighborsClassifier(n_neighbors=1,p=3)
knn_p.fit(x_train, y_train)
print(classification_report(y_test,knn_p.predict(x_test)))
```
